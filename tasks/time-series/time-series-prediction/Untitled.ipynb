```python
import pandas as pd
```

```python
bp_list = []
bv_list = []
av_list = []
ap_list = []
for i in range(1,21):
    str_bp = 'BP'+str(i)
    str_bv = 'BQ'+str(i)
    str_ap = 'AP'+str(i)
    str_av = 'AQ'+str(i)
    bp_list.append(str_bp)
    bv_list.append(str_bv)
    av_list.append(str_av)
    ap_list.append(str_ap)

    
```

```python
df_buy_v = pd.read_csv('buy_v.csv', names=bv_list, header=None)
df_buy_p = pd.read_csv('buy.csv', names=bp_list, header=None)
df_sell_v = pd.read_csv('sell_v.csv', names=av_list, header=None)
df_sell_p = pd.read_csv('sell.csv',  names=ap_list, header=None)
df_event = pd.read_csv('event.csv',  names=['timestamp'], header=None)
```

```python
df = pd.DataFrame()
for i in range(1,21):
    str_bp = 'BP'+str(i)
    str_bv = 'BQ'+str(i)
    str_ap = 'AP'+str(i)
    str_av = 'AQ'+str(i)
    df[str_bp]= df_buy_p[str_bp] 
    df[str_bv]= df_buy_v[str_bv]
    df[str_ap]= df_sell_p[str_ap]
    df[str_av]= df_sell_v[str_av]
    
df['timestamp'] = df_event['timestamp']
```

```python
df.head()
```

```python
df['TP'] = (df['BP1'] + df['AP1'])/2
df['TQ'] = (df['BQ1'] + df['AQ1'])
```

```python
df.head()
```

```python
df.to_csv('amazon_book.csv')
```
